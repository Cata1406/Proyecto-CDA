```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Group the data by 'FACULTAD_PROGRAMA_1' and count unique 'CODIGO_ESTUDIANTE_ANON'
facultad_counts = info_matriculados.groupby('FACULTAD_PROGRAMA_1')['CODIGO_ESTUDIANTE_ANON'].nunique().reset_index()

# Sort by the count of unique 'CODIGO_ESTUDIANTE_ANON'
facultad_counts = facultad_counts.sort_values(by='CODIGO_ESTUDIANTE_ANON', ascending=False)

# PLOT ROOM TYPE DISTRIBUTION BASED ON UNIQUE 'CODIGO_ESTUDIANTE_ANON'
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=facultad_counts, y='FACULTAD_PROGRAMA_1', x='CODIGO_ESTUDIANTE_ANON', order=facultad_counts['FACULTAD_PROGRAMA_1'])
plt.title('Cantidad de estudiantes por facultad')
plt.xlabel('Cantidad de estudiantes')
plt.ylabel('Facultad')

# Add value labels to the bars
for p in ax.patches:
    ax.annotate(f'{p.get_width():.0f}', (p.get_width() + 1, p.get_y() + 0.5), ha='left', va='center')

plt.show()
```

**Insight**
